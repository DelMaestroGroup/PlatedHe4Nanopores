```python
import matplotlib.pyplot as plt
import numpy as np
import colors as col
from numpy import pi as π
from scipy.constants import hbar as ħ
from scipy.constants import k as kb

%config InlineBackend.figure_format = 'retina'
%matplotlib inline
plt.style.use('./include/notebook.mplstyle')
cols = plt.rcParams['axes.prop_cycle'].by_key()['color']
import helium
```

```python
import PIL
data = PIL.Image.open('../Data/Sqw_Sokol_cropped.png')
pix = np.array(data)
```

```python
def ω(m,p,q,KL,qF):
    return (4.0/KL)*np.abs(q/(2*qF) -m + p*(q/(2*qF)-m)**2)
```

## Equation for hard rods

\begin{equation}
\omega_{\pm}(q) = \frac{4}{K_L} \left \lvert \frac{q}{2q_{\rm F}} \pm \left(\frac{q}{2q_{\rm F}}\right)^2\right \rvert
\end{equation}

```python
q = np.linspace(0,5,1000)
plt.plot(q,ω(0,-1,q,1,1.2))
plt.plot(q,ω(0,1,q,1,1.2))
plt.ylim(0,5)
```

## Experimental and Analysis Details
